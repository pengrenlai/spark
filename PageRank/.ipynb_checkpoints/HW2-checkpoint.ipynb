```python
sc.stop()
from pyspark import SparkConf, SparkContext
from operator import add

sc = SparkContext()   

num_node = 10879
beta = 0.8

def computeContribs(urls, rank):
    num_urls = len(urls)
    ret = []
    for url in urls:
        ret.append((url, rank / num_urls))
    return ret

def parsePairs(urls):
    pair = urls.split('\t')
    return (pair[0], pair[1])

text_file = sc.textFile(r"page.txt")
links = text_file.map(parsePairs).groupByKey().cache()
ranks = [1/num_node for i in range(num_node)]

for i in range(20):
    contribs = links.flatMap(lambda x: computeContribs(x[1], ranks[int(x[0])]))
    r_update = contribs.reduceByKey(add).mapValues(lambda x : x*beta).collect()
    
    ranks = [0]*num_node
    S = 0
    for (n, p) in r_update:
        ranks[int(n)] = p
        S += p
    ranks = [r+(1-S)/num_node for r in ranks]
    print('iteration %d finish!' % (i+1))

ranks = [(ranks[i], i) for i in range(len(ranks))]
sorted(ranks, reverse=True)
```
