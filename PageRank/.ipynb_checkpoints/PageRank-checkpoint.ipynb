```python
sc.stop()
from pyspark import SparkConf, SparkContext
from operator import add

conf=SparkConf().set("spark.default.parallelism", 3).set("spark.executor.memory", '4g')
sc = SparkContext(conf=conf)    

num_node = 10879
beta = 0.8

def computeContribs(urls, rank):
    num_urls = len(urls)
    for url in urls:
        yield (url, rank / num_urls)

def parsePairs(urls):
    pair = urls.split('\t')
    return (pair[0], pair[1])

def cleanRank(x):
    if x[1][1] is None:
        return (x[0], x[1][0])
    return (x[0], x[1][1])

text_file = sc.textFile(r"page.txt")
links = text_file.map(lambda x : parsePairs(x)).groupByKey().cache()
ranks = sc.parallelize([(str(i), 1/num_node) for i in range(num_node)], 4)

for i in range(20):
    contribs = links.join(ranks).flatMap(lambda url_urls_rank: computeContribs(url_urls_rank[1][0], url_urls_rank[1][1]))
    ranks_update = contribs.reduceByKey(add).mapValues(lambda rank: rank*beta)
    S = ranks_update.map(lambda x : ('sum', x[1])).reduceByKey(add).collect()[0][1]
    ranks = ranks.mapValues(lambda rank: 0).leftOuterJoin(ranks_update).map(lambda x : cleanRank(x))\
                .mapValues(lambda x : x+((1-S)/num_node))
    print('iteration %d finish!' % (i+1))

sorted(ranks.map(lambda x : (x[1], x[0])).collect(), reverse=True)
```
